**imports**

```python
import numpy as np
import scipy as sp
from scipy import stats
from scipy.stats import norm
from scipy.stats import poisson
import matplotlib.pyplot as plt
```

```python
# make reproducible
np.random.seed(312)
```

**Normal Distribution**

```python
data = np.random.randn(200)
xr_cont = np.linspace(-3,3,100)
plt.hist(data, density = 1) # count but normalized
plt.plot(xr_cont, norm.pdf(xr_cont, loc = 0, scale = 1)) # frequency
plt.title('Normal Distribution')
plt.xlabel("x")
plt.ylabel("frequency");
```

```python
# find features of gaussian distribution
mean, stdev = sp.stats.norm.fit(data)
mean, stdev
```

**Poisson Distribution**

```python
# generate distribution
l = 1
data_2 = np.random.poisson(l , 200)
```

```python
# find mean and stdev under assumption that distribution is normal
mean_2, stdev_2 = sp.stats.norm.fit(data_2)
mean_2, stdev_2
```

```python
# plot sample with normal and poisson shape
xr = np.arange(0,5,1)
plt.hist(data_2, density = 1) # count but normalized
plt.plot(xr, poisson.pmf(xr, mu = 1)) # frequency
plt.plot(xr_cont, norm.pdf(xr_cont, loc = mean_2, scale = stdev_2)) # try fitting normal distribution
plt.title("Poisson Function")
plt.xlabel("x")
plt.xlim([0,4])
plt.legend(["poisson","normal"])
plt.ylabel("frequency");
```
