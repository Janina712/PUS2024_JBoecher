```python
# imports
import pandas as pd
import numpy as np
import pylab as plt
```

```python
# assemble url
repo = "https://github.com/phillydao/phillydao-public-data/blob/main/docs/data/"
my_file = "time_to_disposition_data_daily_by_zip.csv"
qualifiers = "?raw=true"
full_file_name = repo + "/" + my_file  + qualifiers
```

```python
# load data
df_crime = pd.read_csv(full_file_name)
```

```python
# crimes by zip code and ignore date; then get correlation between all features of the df
df_crime.groupby("ZCTA_GEOID").sum().drop("date_value",axis = 1).corr()
```
