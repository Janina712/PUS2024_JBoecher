**4. measures of central tendency**

```
# mean
meanps = pparks["ACREAGE"].mean()
meanps
```

```
# median
medianps = pparks["ACREAGE"].median() # option 1
medianps = pparks["ACREAGE"].quantile() # option 2
medianps
```

```
# are they the same?
meanps == medianps
```

```
print("The mean park size is " + str(round(meanps,2)) + " and the median is " + str(round(medianps,2)) + ".")
print("The statement that the mean and the median are the same is " + str(meanps == medianps) + "!")
```

**Q**: What can you conclude from the values of the mean and median you extracted?


 **A**: The mean is much larger than the median. This may point to outlier data that greatly affect the mean (bring it up), but not so much the median (stays consistent). Most parks have a size "near" 4 acres, while a few parks are very big.  

```
# plot
ax = pparks["ACREAGE"].hist(bins=10) # plot histogram with 10 data bins
ax.set_title("Mean and Median Size of Parks in Philadelphia", weight = 'bold', size = 16) # figure title
ax.set_xlabel("park size in acres") # x-axis title
ax.set_ylabel("number of parks") # y-axis title
ax.axvline(meanps, color = "red") # show mean
ax.axvline(medianps, color = "orange") # show median
ax.legend(["mean", "median"]); # show legend
```


**Figure 1**: This figure depicts a histogram of park size in Philadelphia. Park size measured in terms of acres is on the horizontal axis and the number of parks that are of a certain size is on the vertical axis. The mean and median park size are shown by red and orange vertical bars. We can see that the vast majority of parks (~ 500) fall within the range of zero to ~ 200 acres, while bars of minimal height indicate a very small number of parks of much larger size. The mean (= 20.0 acres) and median (= 3.6 acres) lines further highlight this tendency. The mean is further to the right than the median, indicating the presence of outliers (very big parks) that affect the mean but not the median.

**7. aggregate parks by parent park**

```
pparks.groupby("PARENT_NAME")["ACREAGE"].mean().mean() # group mean
pparks.groupby("PARENT_NAME")["ACREAGE"].mean().median() # group median
```

```
ax = pparks.groupby("PARENT_NAME")["ACREAGE"].mean().hist(bins = 10)
ax.set_title("Size of Parent Parks in Philadelphia", weight = 'bold', size = 16) # figure title
ax.set_xlabel("park size in acres") # x-axis title
ax.set_ylabel("number of parent parks"); # y-axis title
ax.axvline(pparks.groupby("PARENT_NAME")["ACREAGE"].mean().mean(), color = "red") # show mean
ax.axvline(pparks.groupby("PARENT_NAME")["ACREAGE"].mean().median(), color = "orange") # show median
ax.legend(["mean", "median"]); # show legend
```

**Figure 2**: This figure depicts a histogram of average park size by parent park in Philadelphia. Average size of parks that belong to a certain parent park, measured in terms of acres, is on the horizontal axis and the number of parents parks whose parks are, on average, of a certain size is on the vertical axis. The mean (= 9.33) and median (= 3.02) park size are shown by red and orange vertical bars. We can see that the vast majority of parent parks (~ 355) fall within the range of zero to ~ 25 acres, while bars of minimal height indicate a very small number of parents parks of much larger average size. Like before, the mean and median lines suggest the presence of outliers. The mean is further to the right than the median that affect the mean but not the median. However, averaging individual park size within parent park groups reduced the magnitude of outliers. The scope of the x-axis is much reduced and the mean is now closer to the median compared to figure 1.
